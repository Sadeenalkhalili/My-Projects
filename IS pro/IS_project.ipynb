```python
paths = {
    "raw_recipes": "/content/rr.csv",
    "raw_interactions": "/content/ri.csv",
    "preprocessed_recipes": "/content/pp.csv"
}

def load_datasets(paths):
    d = {}

    for name, path in paths.items():
        try:
            if path.endswith(".csv"):
                df = pd.read_csv(path)


            d[name] = df


            print(f"\n📌 **{name.upper()} Dataset:**")
            if isinstance(df, pd.DataFrame):
                print(f"Shape: {df.shape}")
                print(f"Columns: {df.columns.tolist()}")
                print(f"Missing Values:\n{df.isnull().sum()}\n")

            elif isinstance(df, dict):
                print(f"Loaded as a Dictionary. Sample Data:\n{list(df.items())[:5]}")
            else:
                print(f"Loaded as an unknown type. Sample Data: {df}")
            print("-" * 50)

        except Exception as e:
            print(f"❌ Error loading {name}: {e}")

    return d


d= load_datasets(paths)
```

```python
import ast

def clean_data(ds):

    if "raw_recipes" in ds:
        ds["raw_recipes"]["submitted"] = pd.to_datetime(ds["raw_recipes"]["submitted"], errors="coerce")

    if "raw_interactions" in ds:
        ds["raw_interactions"]["date"] = pd.to_datetime(ds["raw_interactions"]["date"], errors="coerce")


    if "raw_recipes" in ds:
        ds["raw_recipes"]["ingredients"] = ds["raw_recipes"]["ingredients"].apply(ast.literal_eval)
        ds["raw_recipes"]["steps"] = ds["raw_recipes"]["steps"].apply(ast.literal_eval)

    print("✅ dataset is cleaned")
    return ds

datasets =clean_data(d)
```

```python
def recipes_with_feedback(case_rpath, interactions_path, output_merged_path):
    case_rdf = pd.read_csv(case_rpath)
    interactions_df = pd.read_csv(interactions_path)

    interactions_df = interactions_df[["recipe_id", "rating", "review"]]

    merged_df = case_rdf.merge(interactions_df, left_on="id", right_on="recipe_id", how="left")

    merged_df.drop(columns=["recipe_id"], inplace=True)

    merged_df.to_csv(output_merged_path, index=False)

    print(f"✅ Recipes successfully merged with user feedback and saved to {output_merged_path}")
    return merged_df


enriched_recipes = recipes_with_feedback(
    case_rpath="/content/case_based_recipes.csv",
    interactions_path="/content/RAW_interactions.csv",
    output_merged_path="enriched_recipes.csv"
)
```

```python
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer

def compute_tfidf(case_rpath, tfidf_vectorizer_path, tfidf_matrix_path):

    cr = pd.read_csv(case_rpath)

    vectorizer = TfidfVectorizer(stop_words='english')
    tm= vectorizer.fit_transform(cr["formatted_ingredients"])


    with open(tfidf_vectorizer_path, 'wb') as file:
        pickle.dump(vectorizer, file)


    with open(tfidf_matrix_path, 'wb') as file:
        pickle.dump(tm, file)

    print(f"✅ TF-IDF vectorizer and matrix saved")
    return vectorizer, tm

compute_tfidf(
    case_rpath="/content/final_case_recipes.csv",
    tfidf_vectorizer_path="tfidf_vectorizer.pkl",
    tfidf_matrix_path="tfidf_matrix.pkl"
)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def compute_similarity(tfidf_matrix, similarity_matrix_path):

    with open(tfidf_matrix, 'rb') as file:
        tfidf_matrix = pickle.load(file)

    similarity_matrix = cosine_similarity(tfidf_matrix)

    with open(similarity_matrix_path, 'wb') as file:
        pickle.dump(similarity_matrix, file)

    print(f"✅ Cosine similarity matrix saved")
    return similarity_matrix

compute_similarity(
    tfidf_matrix="/content/tfidf_matrix.pkl",
    similarity_matrix_path="recipe_similarity.pkl"
)

```

```python
def recipes_by_ingredients(ingredients, case_rpath, tfidf_vectorizer_path, tfidf_matrix_path, similarity_matrix_path, top_n=5):

    cr= pd.read_csv(case_rpath)

    with open(tfidf_vectorizer_path, 'rb') as file:
        vectorizer = pickle.load(file)


    with open(tfidf_matrix_path, 'rb') as file:
        tfidf_matrix = pickle.load(file)


    with open(similarity_matrix_path, 'rb') as file:
        similarity_matrix = pickle.load(file)


    user_tfidf = vectorizer.transform([ingredients])


    similarity_scores = cosine_similarity(user_tfidf, tfidf_matrix).flatten()


    similar_recipe_indices = similarity_scores.argsort()[::-1]


    similar_recipes = cr.iloc[similar_recipe_indices]


    similar_recipes = similar_recipes.drop_duplicates(subset=['id']).head(top_n)

    print(f"✅ Top {top_n} distinct recipes based on input ingredients:")
    print(similar_recipes[['id', 'name', 'formatted_ingredients', 'steps']])

    return similar_recipes


ingredients = input("Enter ingredients (comma-separated): ").strip()

similar_recipes = recipes_by_ingredients(
    ingredients=ingredients,
    case_rpath="/content/final_case_recipes.csv",
    tfidf_vectorizer_path="/content/tfidf_vectorizer.pkl",
    tfidf_matrix_path="/content/tfidf_matrix.pkl",
    similarity_matrix_path="recipe_similarity.pkl",
    top_n=5
)
```

```python
def find_recipes_by_ingredients(ingredients, case_rpath, tfidf_vectorizer_path, tfidf_matrix_path, similarity_matrix_path, top_n=5):

    cr= pd.read_csv(case_rpath)

    with open(tfidf_vectorizer_path, 'rb') as file:
        vectorizer = pickle.load(file)


    with open(tfidf_matrix_path, 'rb') as file:
        tfidf_matrix = pickle.load(file)


    with open(similarity_matrix_path, 'rb') as file:
        similarity_matrix = pickle.load(file)


    input_words =ingredients.split(",")
    vocab = set(vectorizer.get_feature_names_out())
    invalid_words = [word.strip() for word in input_words if word.strip() not in vocab]

    if invalid_words:
        print(f"❌ The following ingredients were not found in the dataset vocabulary: {', '.join(invalid_words)}")
        return None


    user_tfidf = vectorizer.transform([ingredients])


    similarity_scores = cosine_similarity(user_tfidf, tfidf_matrix).flatten()


    similar_recipe_indices = similarity_scores.argsort()[::-1]


    similar_recipes = cr.iloc[similar_recipe_indices]


    similar_recipes = similar_recipes.drop_duplicates(subset=['id']).head(top_n)

    print(f"\n✅ Top {top_n} distinct recipes based on input ingredients:\n")


    for _, row in similar_recipes.iterrows():
        print(f"🍽️ **Recipe: {row['name']} (ID: {row['id']})**")
        print(f"📌 **Ingredients:** {row['formatted_ingredients']}")


        try:
            steps_list = ast.literal_eval(row['steps'])
        except:
            steps_list = row['steps'].strip("[]").split("', '")

        print("📝 **Steps:**")
        for step_num, step in enumerate(steps_list, start=1):
            print(f"   {step_num}. {step.strip()}")

        print("\n" + "-" * 60 + "\n")

    return similar_recipes


ingredients = input("Enter ingredients (comma-separated): ").strip()

similar_recipes = recipes_by_ingredients(
    ingredients=ingredients,
    case_rpath="/content/final_case_recipes.csv",
    tfidf_vectorizer_path="/content/tfidf_vectorizer.pkl",
    tfidf_matrix_path="/content/tfidf_matrix.pkl",
    similarity_matrix_path="recipe_similarity.pkl",
    top_n=5
)
```
